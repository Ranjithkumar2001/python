## Linear Regression

```python
from sklearn.linear_model import LinearRegression
x=np.array([1,2,3,4,5]).reshape((-1,1))
y=np.array([4,12,28,52,80])
plt.scatter(x,y);
plt.show()
```

```python
sur=LinearRegression().fit(x,y)
als=sur.score(x,y)
print('coefficient of determination:',als)
print('intercept:',sur.intercept_)
print('slope:',sur.coef_)
```

## Afinn

```python
from afinn import Afinn
nan = Afinn()
al= nan.score('worst')
su=nan.score('best')
print(al)
print(su)
```

```python
Dict = {}
print("Empty Dictionary: ")
print(Dict)
Dict = {1: 'MCET', 2: 'IT', 3: 'Department'}
print("\nDictionary with the use of Integer Keys: ")
print(Dict)
Dict = {'Name': 'MCET', 1: [1, 2, 3, 4]}
print("\nDictionary with the use of Mixed Keys: ")
print(Dict)
# with dict() method
Dict = dict({1: 'Geeks', 2: 'For', 3:'Geeks'})
print("\nDictionary with the use of dict(): ")
print(Dict)
Dict = dict([(1, 'MCET'), (2, 'IT')])
print("\nDictionary with each item as a pair: ")
print(Dict)
Dict = {1: 'MCET', 'name': 'IT', 3: 'Department'}
print("Accessing a element using key:")
print(Dict['name'])
print("Accessing a element using get:")
print(Dict.get('name'))
```
